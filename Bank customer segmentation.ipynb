# BANK CUSTOMER SEGMENTATION (KMeans Clustering)

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# importing data
data = pd.read_csv('bank_transactions.csv')
data.head()
```

# Processing Data

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.shape
```

```python
data.nunique()
```

```python
data.info()
```

```python
#Change name of columns for clarification

data.rename({'CustomerDOB':'dateofbirth' , 'CustGender' : 'gender' , 'CustLocation' : 'location' , 
           'CustAccountBalance' : 'accountbalance','TransactionAmount (INR)' : 'transactionamount'} , inplace = True , axis = 1)
data.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
#Change data type
data['dateofbirth'] = pd.to_datetime(data['dateofbirth'])
data['transactiondate'] = pd.to_datetime(data['transactiondate'])
```

```python
# Drop unnecessary column
data.drop(columns = 'transactionid' , inplace = True , axis = 1)
```

```python
data.head()
```

```python
#Check duplicated values
data.duplicated().sum()
```

```python
# Convert unit of transaction time to minute
data['transactiontime'] = (data['transactiontime'] / 60).astype('int64')
```

```python
# Calculate age of customers
data['age'] = data['transactiondate'].dt.year - data['dateofbirth'].dt.year
```

```python
data.info()
```

# checking for outliers

```python
num_data = data.select_dtypes(include = ['int','float'])
for i in num_data:
    sns.boxplot(data, x=i)
    plt.title(i)
    plt.show()
```

```python
# seems like there are outliers in age
data = data[(data['age']>=0) & (data['age']<=50)]
```

# Visualization

```python
fig = plt.figure(figsize = (10,5))
fig.suptitle('Categorical Features')

ax1 = fig.add_subplot(1,2,1)
loc_10 = data['location'].value_counts().to_frame().head(10).reset_index()
ax1 = sns.barplot(data = loc_10 , x = 'location' , y = 'index' )
ax1.set_title('Top Ten Locations of Customers')
ax1.bar_label(ax1.containers[0])
ax1.set_ylabel(' ')
ax1.set_xlabel('Location')

ax2 = fig.add_subplot(1,2,2)
pie = data['gender'].value_counts().to_frame().reset_index()
ax2 = plt.pie(data = pie , x = 'gender' , labels = ['Male','Female'] , autopct = '%1.1f%%' , colors = ['blue','orange'],
             wedgeprops={"edgecolor":"k",'linewidth': 2, 'antialiased': True})

plt.title('Gender of Customer')
plt.tight_layout()
```

```python
feature = ['transactiontime','transactionamount','accountbalance']
fig = plt.figure(figsize = (10,3))
fig.suptitle('Numerical Features')

for i,j in enumerate(feature):
    ax = fig.add_subplot(1,3,i+1)
    ax = sns.boxplot(data , x = j)
    ax.set_title(j)
    ax.set_ylabel(' ')
    

plt.tight_layout()
```

```python
# Encode gender
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_columns = ['accountbalance', 'transactiontime']

scaled_data = data.copy()

for i in numerical_columns:
    scaled_data[i] = scaler.fit_transform(scaled_data[[i]])
```

```python
scaled_data = scaled_data[['accountbalance','transactiontime','age']]
scaled_data.head(2)
```

# clustering

```python
from sklearn.cluster import KMeans

inertia = []
num_clusters = list(range(1,10))

for k in num_clusters:
    kmeans = KMeans(n_clusters = k, n_init= 10)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
#Plot the graph
plt.plot(num_clusters , inertia , 'o-')    
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
```

```python
# Find the elbow spot
from kneed import KneeLocator
knee = KneeLocator(x = num_clusters , y = inertia , S = 1.0 , curve = 'convex' , direction = 'decreasing' , online = True)
elbow = knee.elbow
print('Elbow at k = {} clusters'.format(elbow))
```

```python
#Plot the graph

plt.plot(num_clusters , inertia , 'o-')    
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')

plt.axvline(x=elbow, color='black')
```

```python
model = KMeans(n_clusters = 3, n_init=10)
model.fit(scaled_data)
```

```python
model.cluster_centers_.round(1)
```

```python
model.labels_
```

```python
data['cluster'] = model.labels_
```

```python
sns.countplot(data, x='cluster')
```

```python
df1 = data[data.cluster == 0]
df2 = data[data.cluster == 1]
df3 = data[data.cluster == 2]
```

```python
plt.scatter(df1.age,df1['accountbalance'],color='green')
plt.scatter(df2.age,df2['accountbalance'],color='red')
plt.scatter(df3.age,df3['accountbalance'],color='black')
plt.xlabel('age')
plt.ylabel('Balance')
```

```python
mean = data.groupby('cluster')['accountbalance','transactiontime','transactionamount','age'].mean()
fig = plt.figure(figsize = (10,5))
sns.heatmap(data = mean , cmap = 'Blues' , annot = True, fmt = ".1f" )
```

```python
median = data.groupby('cluster')['accountbalance','transactiontime','transactionamount','age'].median()
fig = plt.figure(figsize = (10,5))
sns.heatmap(data = median , cmap = 'Blues' , annot = True, fmt = ".1f" )
```

# *conclusion*


Seems like median is a better measure of clusters. With median it is evident that,

**1.** The cluster 0 represents the customers around age 25, They spend around Rs.287 per transaction and spend total time in application for 2740 minutes, They possess Rs.10553.2 in their accounts.

**2.** The cluster 1 represents the customer around age 32, They spend around Rs.500 per transaction and spend total time in application for 2760 minutes, They possess Rs.20826.7 in their accounts.

**3.** The cluster 2 represents the customer around age 42, They spend around Rs.800 per transaction and spend total time in application for 2737 minutes, They possess Rs.35881.3 in their accounts.

**# In other words customers in cluster-1 are young and they do not spend much per transaction, possess less money in account. So they can be focused for availing "*credit cards*".**

**# Then the customers in cluster-2 are still young but they possess good amount of money as compared to customers in cluster-1, thus they can be targeted for availing "*insurance*".**

**# Whereas the customers in cluster-3 are old and possess lot of money in their accounts, thus they can be targeted either for availing "*trips and tours*" or availing investments for next generations like "*mutual funds*".**
